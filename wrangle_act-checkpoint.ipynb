```python
# this is to actually access twitter to gather data through api
error_list = [] # this is the list of error tweets just in case
df_list = [] # this is the list of tweets for valid ones

for tweet_id in twitter_archive['tweet_id']:
    try:        
        tweet = api.get_status(tweet_id,tweet_mode = 'extended', wait_on_rate_limit = True, wait_on_rate_limit_notify = True, sleep_on_rate_limit=False)
        tweet_json = json.loads(json.dumps(tweet._json))
        favorites = tweet_json['favorite_count'] # How many favorites the tweet had
        retweets = tweet_json['retweet_count'] # Count of the retweet
        
        df_list.append({'tweet_id': int(tweet_id),
                        'favorites': int(favorites),
                        'retweets': int(retweets)})
        
    except Exception as e:
        print(str(tweet_id)+ " " + str(e))
        error_list.append(tweet_id)
        
```

### Quality Issue 6: Clean up rating_numerator and rating demoninator
Some additional assessment will be done on the rating numerator and denominator, by looking at value_counts(), and there is some outliers such as 1776. And I also spot check a few tweets online, and compare to text column, and they don’t match due to picking up the wrong numerator, so I will re-pull all the ratings again from “text” column. 

```python
# check the unique value of numerator and denominator first:
tweet_clean['rating_numerator'].value_counts()
```

```python
tweet_clean['rating_denominator'].value_counts()
```

```python
# this is to run one line of code to check what would come back if trying to pull the ratings from the text. 
m = re.findall(r'(\d+(\.\d+)|(\d+))\/(\d+0)',
        "Two dogs in this one. Both are rare Jujitsu Pythagoreans. One slightly whiter than other. Long legs. 7/10 and 8/10")
m
```

```python
#it would be very manual to clean up all the numerator and denominator, I can repull the data and then clean up only exceptions
#if neccessary
#len of score means how many list it has in the text field. so there are several situations here. 
# such as 1 list, 2 lists and no list.
numerator = []
denominator = []
for text in tweet_clean['text']:
    score = re.findall(r'(\d+(\.\d+)|(\d+))\/(\d+0)',text)
    if len(score) == 0:
        numerator.append('NaN')
        denominator.append('NaN')
    elif len(score) == 1:
        n = score[0][0]
        m = score[0][-1]
        numerator.append(n)
        denominator.append(m)
    elif len(score) > 1 and score[0][-1] == '10':
        n = score[0][0]
        m = score[0][-1]
        numerator.append(n)
        denominator.append(m)
    elif len(score) > 1 and score[1][-1] == '10':
        n = score[1][0]
        m = score[1][-1]
        numerator.append(n)
        denominator.append(m)
    else:
        numerator.append('0')
        denominator.append('0')
tweet_clean["rating_numerator"] = numerator
tweet_clean["rating_denominator"] = denominator
```

```python
#check the record with 0 numerator,and it looks like all these data pulls below are correct. 
pd.set_option('display.max_colwidth',-1)
tweet_clean[tweet_clean.rating_numerator == 0 ][['tweet_id','text','rating_numerator','rating_denominator']]
```

### Quality Issue 8: repull dog names for the values don't look like a name
There were 546 “None” values under Name column, and 55 “a”, and there were also some odd values such as “the”, “this”, etc. so I re-pulled the names from the text column, after observing the patterns in Excel. 

```python
# I pulled the value and counts for each names, and also checked the names in Excel to get a better visualization.
tweet_clean['name'].value_counts()
```

```python
# repull all the names as below: 
new_name = []
for text in tweet_clean['text']:

    # This is "***", dog's name is index 2->3rd place
    if text.startswith("This is") and re.match(r'[A-Z].*',text.split()[2]):
        new_name.append(text.split()[2].strip(".").strip(","))
        
    # Here is "***", dog's name is index 2->3rd place
    elif text.startswith("Here is") and re.match(r'[A-Z].*',text.split()[2]):
        new_name.append(text.split()[2].strip(".").strip(","))
        
    # Meet "***", dog's name index 1->2nd place
    elif text.startswith("Meet") and re.match(r'[A-Z].*',text.split()[1]):
        new_name.append(text.split()[1].strip(".").strip(","))
        
    #Say hello to "***", dog's name index 3->4th place
    elif text.startswith("Say hello to") and re.match(r'[A-Z].*',text.split()[3]):
        new_name.append(text.split()[3].strip(".").strip(","))
        
    # Here we have "***", dog's name index 3->4th place
    elif text.startswith("Here we have") and re.match(r'[A-Z].*',text.split()[3]):
        new_name.append(text.split()[3].strip(".").strip(","))
        
    # after "named" "***"
    elif "named" in text and re.match(r'[A-Z].*',text.split()[text.split().index("named")+1]):
        new_name.append(text.split()[text.split().index("named")+1].strip(".").strip(","))

    #to extract name after "name is" "***".
    elif "name is" in text and re.match(r'[A-Z].*',text.split('name is')[1].split()[0]):
            new_name.append(text.split('name is')[1].split()[0].strip(".").strip(","))
    else:
        new_name.append("NaN")
new_name
```

```python
# this is to check how many names we have in new_name, it should be 1994, so no nulls.
len(new_name)
```

```python
# to override the old column
tweet_clean['name'] = new_name
```
